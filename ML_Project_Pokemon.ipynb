```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon=pd.read_csv('/content/drive/MyDrive/kaggle Files/pokemon.csv',na_values={'height_m':[" ","n.a"],'percentage_male':[" ","n.a"],
                                                                                 'type2':[" ","n.a"],'weight_kg':[" ","n.a"]})
pokemon
```

```python
sns.heatmap(pokemon.isnull(),yticklabels=False,cbar=False)
```

```python
pokemon.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
pokemon['abilities']=le.fit_transform(pokemon['abilities'])
pokemon['capture_rate']=le.fit_transform(pokemon['capture_rate'])
pokemon['classfication']=le.fit_transform(pokemon['classfication'])
pokemon['japanese_name']=le.fit_transform(pokemon['japanese_name'])
pokemon['name']=le.fit_transform(pokemon['name'])
pokemon['type1']=le.fit_transform(pokemon['type1'])
pokemon['type2']=le.fit_transform(pokemon['type2'])
```

```python
pokemon['height_m'].fillna(pokemon['height_m'].median(), inplace=True)
pokemon['percentage_male'].fillna(pokemon['percentage_male'].median(), inplace=True)
pokemon['type2'].fillna(pokemon['type2'].median(), inplace=True)
pokemon['weight_kg'].fillna(pokemon['weight_kg'].median(), inplace=True)
```

```python
pokemon.isnull().sum()
```

```python
x=pokemon.drop(['is_legendary'],axis=1)
x.head(3)
```

```python
y=pokemon.iloc[:,-1].values
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
y_pred_lr=lr.predict(x_test)
y_pred_lr
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred_lr))
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm_lr=confusion_matrix(y_test,y_pred_lr)
cm_lr
```
